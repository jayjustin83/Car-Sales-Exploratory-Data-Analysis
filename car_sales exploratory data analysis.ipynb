```python
# changing the latest_launch datatype
df.Latest_Launch = df.Latest_Launch.astype('datetime64[ns]')
```

```python
# QUESTION 10: Relationship between price and latest launch year
sns.lineplot(data=df, x='Latest_Launch', y='Price_in_thousands', marker='o')
plt.title('Price_in_thousands vs. Latest_Launch')
plt.xlabel('Latest_Launch')
plt.ylabel('Price_in_thousands')
plt.grid(True)
plt.show()
```

```python
# slicing for the year 2010 to 2013 for more exploration.
after_2010 = df[df['Latest_Launch'] > '01-01-2011']
after_2010
```

```python
# Trend for price and latest launch year for 2011 to 2013 for more exploration.
sns.lineplot(data=after_2010, x='Latest_Launch', y='Price_in_thousands', marker='o')
plt.title('Price_in_thousands vs. Latest_Launch')
plt.xlabel('Latest_Launch')
plt.ylabel('Price_in_thousands')
plt.grid(True)
plt.show()
```

```python
# QUESTION 11: Group car by the maufacturer and analyze their price and fuel efficiency
grouped_data = df.groupby('Manufacturer').agg({
    'Price_in_thousands': 'mean',
    'Fuel_efficiency': 'mean'
}).reset_index()
grouped_data
```

```python
# visualizing the realationship between price_in_thousand and manufacturer
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Manufacturer', y='Price_in_thousands', color='blue')
plt.title('Price_in_thousands vs. Manufacturers')
plt.xlabel('Manufacturers')
plt.ylabel('Price_in_thousands')
plt.xticks(rotation=90)
plt.show()
```

```python
# visualizing the relationship between fuel efficiency and manufacturer
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Manufacturer', y='Fuel_efficiency', color='red')
plt.title('Average Fuel Efficiency by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Fuel Efficiency')
plt.xticks(rotation=90)
plt.show()
```
